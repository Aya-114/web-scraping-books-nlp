```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

base_url = "https://books.toscrape.com/catalogue/page-{}.html"
book_base_url = "https://books.toscrape.com/catalogue/"
```

```python


all_books = []

for page in range(1, 51):
    print(f"Scraping page {page}...")
    
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "lxml")
    
    books = soup.find_all("article", class_="product_pod")
    
    for book in books:
        
        # Book name
        title = book.h3.a["title"]
        
        # Price
        price = book.find("p", class_="price_color").text.strip()
        
        # Stock availability
        availability = book.find("p", class_="instock availability").text.strip()
        
        # Book page link
        relative_link = book.h3.a["href"]
        book_url = book_base_url + relative_link
        
        # Open book page to get description
        book_response = requests.get(book_url)
        book_soup = BeautifulSoup(book_response.text, "lxml")
        
        description = book_soup.find("meta", attrs={"name": "description"})
        if description:
            description = description["content"].strip()
        else:
            description = "No description"
        
        all_books.append({
            "Page": page,
            "Title": title,
            "Price": price,
            "Availability": availability,
            "Description": description
        })
```

```python

# Save to CSV
df = pd.DataFrame(all_books)
df.to_csv("books_scraped.csv", index=False)

print("Scraping Finished ✅")
```
